```
!pip install -q tabula-py
import tabula
```

```
pdf_path = "/content/Scope of work for Oasys Trade Settlement Documentation Project.pdf"

dfs = tabula.read_pdf(pdf_path,pages=1, stream=True)
# read_pdf returns list of DataFrames
#print(len(dfs))
dfs[0]
```

```
headers = [c.replace(' ', '_') for c in dfs[0].columns]
#print(headers)
dfs[0].columns = [c.replace(' ', '_') for c in dfs[0].columns]
#print ('\n--------\n')
#print(dfs[0].columns)
```

```
dfs[0]["grp"]=(~dfs[0][headers[0]].isna()).cumsum()
```

```
for i in headers:
  dfs[0][i]=dfs[0][i].astype(str)
```

```
sampdict = {}
for i in headers:
  sampdict[i] = " ".join

#samp = {"S.no": " ".join, "External_System": " ".join, "Description": " ".join, "Interface_type": " ".join}

dfs[0]=dfs[0].groupby("grp").agg(sampdict).reset_index(drop=True)
dfs[0]
```

```
for i in headers:
  dfs[0][i] = dfs[0][i].replace({'nan': ''}, regex=True)

dfs[0].to_excel("output.xlsx",sheet_name='page_1')  
```

Installing PyPDF2 to count the number of pages in PDF


```
!pip install PyPDF2
```

```
from PyPDF2 import PdfFileReader

with open("/content/Scope of work for Oasys Trade Settlement Documentation Project.pdf", "rb") as pdf_file:

  pdf_reader = PdfFileReader(pdf_file)

  numofPages = pdf_reader.numPages

  print(f"The total number of pages in the pdf document is {numofPages}")
```

```
# You can convert from pdf into JSON, CSV, TSV
for i in range(1,numofPages):
  tabula.convert_into(pdf_path, f"table-in-page-{i}.csv", pages=i, output_format="csv")
  !cat f"table-in-page-{i}".csv
```
